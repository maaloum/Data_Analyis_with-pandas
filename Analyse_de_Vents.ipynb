### Fusionner toutes les mois en un seul 

```python
path = 'C:\Users\Maaloum\Desktop\Pandas-Data-Science-Tasks-master\SalesAnalysis\Sales_Data'
files = [file for file in os.listdir(path) if not file.startswith('.')]
All_months = pd.DataFrame()
for file in files :
    current_data = pd.read_csv(path+"/"+file)
    All_months = pd.concat([All_months, current_data]) 
All_months.to_csv("Tous_les mois.csv", index=False)
All_months.head(5)
```

Relire les nouvelles données

```python
tt_data = pd.read_csv("Tous_les mois.csv")
tt_data.head()
```

### Clean up the data!
cette étape consiste à supprimer les données indéserables ou manquantes afin de rendre les données plus sgnificatives


```python
tt_data = tt_data.replace(' ?', np.nan).dropna()
#display(tt_data.head())
tt_data.head()
```

### Feature Engeneering 

```python
#Ajouter une colonne de mois 
```

```python
tt_data = tt_data[tt_data['Order Date'].str[0:2]!='Or']
tt_data.tail()
```

```python
tt_data['Month'] = tt_data['Order Date'].str[0:2]
tt_data['Month'] = tt_data['Month'].astype('int32')
tt_data.tail()
```

```python
#Ajour une col pour les villes 
```

```python
def get_city(address):
    return address.split(",")[1].strip(" ")

def get_state(address):
    return address.split(",")[2].split(" ")[1]

tt_data['City'] = tt_data['Purchase Address'].apply(lambda x: get_city(x))
tt_data.head()
```

```python
 #convertir les colonne en numeric 
```

```python
tt_data['Quantity Ordered'] = pd.to_numeric(tt_data['Quantity Ordered'], errors='coerce')
tt_data['Price Each'] = pd.to_numeric(tt_data['Price Each'], errors='coerce')
tt_data['Quantity Ordered'].dtype
```

### Data Exploration

#Quesion 1 : Quel est le meilleur mois de vent ? et combien y'a t-il de gains

```python
tt_data['Sales'] = tt_data['Quantity Ordered'] * tt_data['Price Each'].astype('float')
tt_data.tail()
```

```python
sales_month = tt_data.groupby(['Month']).sum()['Sales']
```

```python
months = range(1,13)
print(months)

plt.bar(months,sales_month)
plt.xticks(months)
plt.ylabel('Sales in USD ($)')
plt.xlabel('Month number')
plt.show()
```
